```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("demand_gendata.csv")
df.head()
```

```python
df.groupby(['month','Product']).sum()
```

```python
# Group by 'month' and 'Product', sum the 'demand' for each product in each month
profitable_category = pd.DataFrame(df.groupby(['month', 'Product']).sum()['demand'])
profitable_category
```

```python
# Sort the DataFrame by 'demand' in descending order
profitable_category.sort_values(by=['month', 'demand'], inplace=True, ascending=[True, False])
profitable_category
```

```python
# Display the highly demanded products for each month

profitable_category.reset_index(inplace=True)
profitable_category
```

```python
def plot_sales_details(selected_month):
    # Filter data for the selected month
    selected_data = df[df['month'] == selected_month]
    # Create a bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Product', y='demand', data=selected_data, palette='viridis')
    plt.title(f'Sales Details for {selected_month}')
    plt.xlabel('Product')
    plt.ylabel('Demand')
    plt.show()
    # Identify highly sold and lowly sold products
    highly_sold_product = selected_data.loc[selected_data['demand'].idxmax()]['Product']
    lowly_sold_product = selected_data.loc[selected_data['demand'].idxmin()]['Product']

    print(f'Highly sold product in {selected_month}: {highly_sold_product}')
    print(f'Lowly sold product in {selected_month}: {lowly_sold_product}')


# Example: Plot sales details for February
plot_sales_details('April')
```
